```python
# Importing the libraries
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# Reading the data
url = "https://archive.ics.uci.edu/static/public/350/data.csv"
df = pd.read_csv(url, sep= ',')
```

```python
# Generating the Profile Report
default_report = ProfileReport(df, title="EDA Report for Default of Credit Card Clients")
default_report.to_file(output_file = 'EDA Report for Default of Credit Card Clients.html')
```
